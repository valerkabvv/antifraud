```python
from sklearn.model_selection import train_test_split
```

```python
def tr_tst_split(tr):
    genue = tr[tr.resolution == "G"].sort_values("rec_user_id")
    fraud = tr[tr.resolution == "F"].sort_values("rec_user_id")
    
    fraud_tr, fraud_tst = train_test_split(fraud, test_size = 0.47, shuffle = False)
    genue_tr, genue_tst = train_test_split(genue, test_size = 0.47, shuffle = False)
    print(set(fraud_tr.rec_user_id).intersection(fraud_tst.rec_user_id))
    
    return pd.concat([fraud_tr, genue_tr]), pd.concat([fraud_tst, genue_tst])
```

```python
train, tst = tr_tst_split(tr)
```

### На тестовых данных посчитаю распределение по времени в мошеннеческих транзакциях

```python
def genue_fraud_mu_std(tr_hist, tr):
    genue = tr[tr.resolution == "G"][["user_id", "rec_user_id"]].values.ravel()
    fraud = tr[tr.resolution == "F"].rec_user_id
    
    fr_times = time_features.calculate_hours(tr_hist[tr_hist.user_id.isin(fraud)].date_time)
    gn_times = time_features.calculate_hours(tr_hist[tr_hist.user_id.isin(genue)].date_time)
    
    fr_mu = time_features.calculate_periodic_mean(fr_times)
    fr_std = time_features.calculate_periodic_std(fr_times)
    
    gn_mu = time_features.calculate_periodic_mean(gn_times)
    gn_std = time_features.calculate_periodic_std(gn_times)
    
    return [fr_mu, fr_std, gn_mu, gn_std]
```

```python
fr_mu, fr_std, gn_mu, gn_std = genue_fraud_mu_std(tr_history, tst)
```

```python
intervals_fr = tr_history.groupby("user_id").date_time.agg(lambda x: time_features.is_in_interval(fr_mu, fr_std, x))
intervals_fr = intervals_fr/tr_history.groupby("user_id").date_time.agg("count")

intervals_gn = tr_history.groupby("user_id").date_time.agg(lambda x: time_features.is_in_interval(gn_mu, gn_std, x))
intervals_gn = intervals_gn/tr_history.groupby("user_id").date_time.agg("count")
```

```python
intervals = pd.concat([intervals_fr, intervals_gn], axis =1)
intervals.columns = ["fraud interval", "genue interval"]

intervals
```

```python
us = tr.groupby("user_id").rec_user_id.agg("count")
rec_us = tr.groupby("rec_user_id").event_id.agg("count")
```

```python
count = pd.concat([us, rec_us], axis =1).fillna(0)
count.columns = ["get_tr", "completed_tr"]
count["ratio"] = count.apply(lambda x: x.completed_tr/(x.sum()), axis = 1)
```
